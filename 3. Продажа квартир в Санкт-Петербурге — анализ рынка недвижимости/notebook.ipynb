### **Исправление высоты потолков**

```python
def descr_numeric(target_column):
    print(data[target_column].head(7), data[target_column].describe())
    data.boxplot(column=target_column, figsize=(3,10)); 
    plt.show()
    data[target_column].hist(figsize=(10,5), bins=100);
descr_numeric('ceiling_height')
```

**Что сделал:**
Написал функцию для описания стобцов, понимаю что придется часто пользоваться.Будем считать нормальной высотой потолков  дапазон 250-350 см.

**Что сделал:**
Посмотрел уникальные строки в дата фрейме.

```python
data['ceiling_height'].hist(figsize=(15,4), bins=50, range=(2.3, 4.5));
plt.show()
data['ceiling_height'].hist(figsize=(15,4), bins=50, range=(3.5,35));
```

**Что сделал:**
Вывел для наглядности распределение высот. Считаю, что теория подтвердена, основные высоты потолков сосредоточены в диапазоне 2.5-3.5 метра. Далее посмотрю, что с нестандартными значениями, которые больше 20 и сколько их

```python
kit_min = 5
liv_min = 15
noliv_min = 8

ind_small_noliv = data.loc[ (data['studio']==False) & (((data['total_area'] - data['living_area'] - \
                                                        data['kitchen_area']) < noliv_min) )].index
display(data.loc[ind_small_noliv,['total_area', 'living_area','kitchen_area']])

square_table = data.loc[ind_small_noliv,'total_area'] - data.loc[ind_small_noliv,'living_area'] - \
data.loc[ind_small_noliv,'kitchen_area']
print(square_table)
square_table.describe()
```

**Что сделал:**
Даже присутствуют отрицательные значения. Восполню остальную площадь до минимума (5кв.м) за счет жилой и кухни. Буду урезать каждую на 2%, пока не будут достигнуты минимумы или другая(нежилая) станет около 5 кв.м
Напишу функцию cut_square(ind) которая будет принимать индекс некорректной строки и "урезать" площади, если удачно возвращает 0.

```python
def cut_square(ind):
    kit = data.loc[ind, 'kitchen_area']
    liv = data.loc[ind, 'living_area']
    tot = data.loc[ind, 'total_area']
    kit_min = 5
    liv_min = 15
    noliv_min = 8
    cut = 0.98 
    while noliv_min >= tot - kit - liv:
        if liv > liv_min:
            liv = liv * cut  
        if kit > kit_min:
            kit = kit * cut  
        
    noliv = tot - kit - liv    
    data.loc[ind, 'kitchen_area'] = kit
    data.loc[ind, 'living_area'] = liv
    return [ind, tot, liv, kit, noliv]
    

c=0 
for index in ind_small_noliv:
    print(data.loc[index,['total_area', 'living_area', 'kitchen_area']])
    print(cut_square(index))
    print(data.loc[index,['total_area', 'living_area', 'kitchen_area']])
    print()
    c+=1
print("Произведенено замен =",c)
```
